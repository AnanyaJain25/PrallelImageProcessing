```python
import numpy as np
import matplotlib.pyplot as plt

# Read execution times from file
execution_times = {}
try:
    with open("execution_times.txt", "r") as file:
        for line in file:
            if line.strip():
                parts = line.strip().split(": ")
                if len(parts) == 2:
                    try:
                        execution_times[parts[0]] = float(parts[1].split()[0])
                    except ValueError:
                        print(f"Warning: Invalid time format in line: {line.strip()}")
except FileNotFoundError:
    print("Error: execution_times.txt not found")
    exit(1)

# Define all operations
operations = ["Brightness", "Grayscale", "Blurring", "Sharpening", "Edge Detection"]
serial_times = []
parallel_times = []

# Validate and collect data
for op in operations:
    serial_key = f"{op} Serial"
    parallel_key = f"{op} Parallel"
    if serial_key not in execution_times or parallel_key not in execution_times:
        print(f"Warning: Missing execution times for {op}")
        serial_times.append(0)
        parallel_times.append(0)
    else:
        serial_times.append(execution_times[serial_key])
        parallel_times.append(execution_times[parallel_key])

# Create bar chart comparison
x = np.arange(len(operations))
width = 0.35

plt.figure(figsize=(12, 6))
bars_serial = plt.bar(x - width/2, serial_times, width, label="Serial", color='red')
bars_parallel = plt.bar(x + width/2, parallel_times, width, label="Parallel", color='green')

# Add value labels on top of bars
for bar in bars_serial + bars_parallel:
    height = bar.get_height()
    if height > 0:  # Only annotate non-zero bars
        plt.text(
            bar.get_x() + bar.get_width() / 2, height,
            f'{height:.2f}', ha='center', va='bottom', fontsize=8
        )

# Customize plot
plt.xlabel("Image Processing Operations")
plt.ylabel("Execution Time (ms)")
plt.title("Serial vs. Parallel Execution Time Comparison")
plt.xticks(x, operations, rotation=15)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()

# Optional: Use logarithmic scale if execution times vary widely
# plt.yscale('log')
# plt.ylabel("Execution Time (ms, log scale)")

# Save and show the plot
plt.savefig("execution_time_comparison.png", dpi=300)
plt.show()

# Optional: Plot speedup (Serial Time / Parallel Time)
speedups = [s / p if p > 0 else 0 for s, p in zip(serial_times, parallel_times)]
plt.figure(figsize=(10, 5))
plt.bar(x, speedups, color='blue')
plt.xlabel("Image Processing Operations")
plt.ylabel("Speedup (Serial / Parallel)")
plt.title("Speedup of Parallel vs. Serial Execution")
plt.xticks(x, operations, rotation=15)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.savefig("speedup_comparison.png", dpi=300)
plt.show()
```
